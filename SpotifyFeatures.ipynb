```python
song = df.iloc[10]  # pick one song

recommend = df[
    (df["energy"] > song["energy"] - 0.1) &
    (df["energy"] < song["energy"] + 0.1) &
    (df["danceability"] > song["danceability"] - 0.1) &
    (df["danceability"] < song["danceability"] + 0.1)
]

recommend.head(10)
```
